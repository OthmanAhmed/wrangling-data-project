# Act Report

A lot of question may arise from following the @WeRateDogs twitter account.
**What dog stage the user likes most? What type of dog he likes?**
**Do followers like the same dog stage and type?**
**Is retweet corrponds to favoriting?**
We gathered data from multiple sources cleaned it and explored it to answer these questions.

Let's start by the first question **What dog stage the user likes most?**
The answer is in the following table that deomnstrates the average rating he gave for every dog stage. He gaves nearly average rating to all dog stages except for *Puppers*. That's exciting!.
The bar chart after table may give a sense for that conclusion.

| Dog Stage     | Doggo | Floofer | Pupper | Puppo |
|---------------|-------|---------|--------|-------|
| Mean Rating   | 1.18  | 1.2     | 1.07   | 1.2   |
| Mean Favorite | 18217 | 11906   | 6636   | 20712 |
| Mean retweet  | 6205  | 4286    | 2074   | 5733  |


```python
import pandas as pd 
from matplotlib import pyplot as plt

df = pd.read_csv('twitter_archive_master.csv')
df.groupby('dog_stage')['rating_fraction'].mean().plot(kind='bar');
plt.xlabel('Dog Stage')
plt.ylabel('Rating Fraction')
plt.title('Dog Stages Rating Fractions')
df.groupby('dog_stage')['rating_fraction'].mean()
```

**Do followers like the same dog stages?** Apparently, yes from the previous table. Although, they may not prefer floofers as he does. They also may not prefer puppers. Defenitely followers like *Puppos*. Look at following chart.

```python
df.groupby('dog_stage')['favorite_count'].mean().plot(kind='bar');
plt.xlabel('Dog Stage')
plt.ylabel('Favorite Count')
plt.title('Dog Stages Favorite Count')
df.groupby('dog_stage')['favorite_count'].mean()
```

**Is retweeting corresponds to favoriting?** From previous table and the following chart we find that to some extent favoriting may relate to retweeting.
